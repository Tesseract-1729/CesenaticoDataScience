```python
import numpy as np
import pandas as pd
```

# 2003

```python
df_03 = pd.read_csv("data/datiSporchi/cese03.csv", encoding="utf8")
df_03.head(9)
```

```python
# Questo torna utile in tutti gli anni

# Questa funzione restituisce la classe (numero) a partire dalla stringa
def classe(str):
    if str.startswith("I "):
        return 1
    elif str.startswith("II "):
        return 2
    elif str.startswith("III "):
        return 3
    elif str.startswith("IV "):
        return 4
    elif str.startswith("V "):
        return 5
    else:
        return np.nan

def medaglia(str):
    if str == "oro":
        return "ORO"
    elif str == "argento":
        return "ARGENTO"
    elif str == "bronzo":
        return "BRONZO"
    elif str == "menzione":
        return "MENZIONE"
    else:
        return "-"

df_vuoto = pd.DataFrame(columns=["anno", "studente", "posizione", "regione", "classe", "es1", "es2", "es3", "es4", "es5", "es6", "tot", "medaglia"])
df_vuoto
```

```python
cese_03 = df_vuoto
```

```python
num_03 = 297

for i in range(num_03):
    cese_03.at[i, "anno"] = 2003
    
    cese_03.at[i, "studente"] = df_03.at[i*3, "studente"]
    # La posizione si sistema più avanti
    cese_03.at[i, "regione"] = df_03.at[i*3+1, "distretto"]
    cese_03.at[i, "classe"] = classe( df_03.at[i*3+2, "studente"] )
    
    cese_03.at[i, "es1"] = int( df_03.at[i*3, "es 1"] )
    cese_03.at[i, "es2"] = int( df_03.at[i*3, "es 2"] )
    cese_03.at[i, "es3"] = int( df_03.at[i*3, "es 3"] )
    cese_03.at[i, "es4"] = int( df_03.at[i*3, "es 4"] )
    cese_03.at[i, "es5"] = int( df_03.at[i*3, "es 5"] )
    cese_03.at[i, "es6"] = int( df_03.at[i*3, "es 6"] )

    cese_03.at[i, "tot"] = int( df_03.at[i*3, "tot"] )
    cese_03.at[i, "medaglia"] = medaglia( df_03.at[i*3, "medaglia"] )
```

```python
cese_03
```

```python
cese_03 = cese_03.astype({"anno": int, "studente": str, "posizione": float, "regione": str, "classe": int,
                          "es1": int, "es2": int, "es3": int, "es4": int, "es5": int, "es6": int,
                          "tot": int, "medaglia": str})
```

```python
cese_03.dtypes
```

```python
cese_03 = cese_03.sort_values("tot", ascending=False)
cese_03
```

```python
cese_03.at[0, "posizione"] = 1
for i in range(1, num_03):
    if cese_03.at[i, "tot"] == cese_03.at[i-1, "tot"]:
        cese_03.at[i, "posizione"] = cese_03.at[i-1, "posizione"]
    else:
        cese_03.at[i, "posizione"] = i+1

cese_03 = cese_03.astype({"posizione": int})
```
