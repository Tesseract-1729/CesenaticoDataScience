```python
def verifica_autenticita(x):
    vista = Cese[Cese["studente"]==x]
    vista_anno = vista["anno"].reset_index(drop=True)
    bool01 = vista_anno[1] == vista_anno[0] + 1
    bool12 = vista_anno[2] == vista_anno[1] + 1
    bool23 = vista_anno[3] == vista_anno[2] + 1
    bool34 = vista_anno[4] == vista_anno[3] + 1
    return bool01 & bool12 & bool23 & bool34
```

```python
verifica_autenticita("Bertoli Michele")
```

```python
for x in a.index:
    if not verifica_autenticita(x):
        print(x, "non è passato davvero 5 volte")
```

```python
df = pd.DataFrame(columns=["studente", "anno_finale",
                           "pos1", "pos2", "pos3", "pos4", "pos5",
                           "pt1", "pt2", "pt3", "pt4", "pt5",
                           "menzione", "bronzo", "argento", "oro", "medaglie"])
df
```

```python
for i in range( len(a) ):
    nome = a.index[i]
    vista = Cese[ Cese["studente"] == nome ].reset_index(drop=True)
    
    df.at[i, "studente"] = nome
    df.at[i, "anno_finale"] = vista.at[4, "anno"]
    
    df.at[i, "pos1"] = vista.at[0, "posizione"]
    df.at[i, "pos2"] = vista.at[1, "posizione"]
    df.at[i, "pos3"] = vista.at[2, "posizione"]
    df.at[i, "pos4"] = vista.at[3, "posizione"]
    df.at[i, "pos5"] = vista.at[4, "posizione"]
    
    df.at[i, "pt1"] = vista.at[0, "tot"]
    df.at[i, "pt2"] = vista.at[1, "tot"]
    df.at[i, "pt3"] = vista.at[2, "tot"]
    df.at[i, "pt4"] = vista.at[3, "tot"]
    df.at[i, "pt5"] = vista.at[4, "tot"]

    menzione = 0
    bronzo = 0
    argento = 0
    oro = 0
    for j in range(5):
        if vista.at[j, "medaglia"] == "MENZIONE":
            menzione += 1
        elif vista.at[j, "medaglia"] == "BRONZO":
            bronzo += 1
        elif vista.at[j, "medaglia"] == "ARGENTO":
            argento += 1
        elif vista.at[j, "medaglia"] == "ORO":
            oro += 1
    df.at[i, "menzione"] = menzione
    df.at[i, "bronzo"] = bronzo
    df.at[i, "argento"] = argento
    df.at[i, "oro"] = oro
    df.at[i, "medaglie"] = bronzo + argento + oro

df
```

```python
df = df.sort_values("anno_finale", ascending=True)
df
```

```python
df = df.reset_index(drop=True)
```

```python
df = df.astype({"studente": str, "anno_finale": int,
              "pos1": int, "pos2": int, "pos3": int, "pos4": int, "pos5": int,
              "pt1": int, "pt2": int, "pt3": int, "pt4": int, "pt5": int,
              "menzione": int, "bronzo": int, "argento": int, "oro": int, "medaglie": int})
df.dtypes
```
