```python
import numpy as np
import pandas as pd
```

```python
def assegna_menzioni(df):
    mask_1 = df["es1"] == 7
    mask_2 = df["es2"] == 7
    mask_3 = df["es3"] == 7
    mask_4 = df["es4"] == 7
    mask_5 = df["es5"] == 7
    mask_6 = df["es6"] == 7
    mask_a = df["medaglia"] == "-"
    mask_papabile = mask_1 | mask_2 | mask_3 | mask_4 | mask_5 | mask_6
    mask_tot = mask_papabile & mask_a

    i = 0
    for b in mask_tot:
        if b:
            df.at[i, "medaglia"] = "MENZIONE"
        i = i+1

    return df
```

```python
df_vuoto = pd.DataFrame(columns=["anno", "studente", "posizione", "regione", "classe", "es1", "es2", "es3", "es4", "es5", "es6", "tot", "medaglia"])
df_vuoto
```

```python
def medaglia(tot, b, a, o):
    if tot >= o:
        return "ORO"
    elif tot >= a:
        return "ARGENTO"
    elif tot >= b:
        return "BRONZO"
    else:
        return "-"
```

```python
def pulisci_df_1(anno, xx, numPartecipanti, b, a, o):
    
    # Prendo i due dataframe
    df = pd.read_csv("data/datiSporchi/cese"+xx+".csv", encoding="utf8")
    cese = df_vuoto

    # Inizio a popolare quello bello
    for i in range(numPartecipanti):
        cese.at[i, "anno"] = anno

        cese.at[i, "studente"] = df.at[i, "Cognome"] + " " + df.at[i, "Nome"]
        # La posizione si sistema pi√π avanti
        cese.at[i, "regione"] = "()"
        cese.at[i, "classe"] = -1

        cese.at[i, "es1"] = int( df.at[i, "Esercizio 1"] )
        cese.at[i, "es2"] = int( df.at[i, "Esercizio 2"] )
        cese.at[i, "es3"] = int( df.at[i, "Esercizio 3"] )
        cese.at[i, "es4"] = int( df.at[i, "Esercizio 4"] )
        cese.at[i, "es5"] = int( df.at[i, "Esercizio 5"] )
        cese.at[i, "es6"] = int( df.at[i, "Esercizio 6"] )

        cese.at[i, "tot"] = int( df.at[i, "Totale"] )
        cese.at[i, "medaglia"] = medaglia( df.at[i, "Totale"], b, a, o )

    # Cambio il tipo dei dati
    cese = cese.astype({"anno": int, "studente": str, "posizione": float, "regione": str, "classe": int,
                        "es1": int, "es2": int, "es3": int, "es4": int, "es5": int, "es6": int,
                        "tot": int, "medaglia": str})

    # Ordino i dati per punteggio
    cese = cese.sort_values("tot", ascending=False)

    # Sistemo l'indice
    cese = cese.reset_index(drop=True)

    # Popolo la colonna della posizione in classifica
    pos = 1
    cese.at[0, "posizione"] = pos
    for i in range(1, len(cese.index)):
        if cese.at[i, "tot"] == cese.at[i-1, "tot"]:
            #cese.at[i, "posizione"] = cese.at[i-1, "posizione"]
            cese.at[i, "posizione"] = pos
        else:
            cese.at[i, "posizione"] = i+1
            pos = i+1

    cese = cese.astype({"posizione": int})

    # Assegno le menzioni
    cese = assegna_menzioni(cese)

    # Salvo il database su pickle
    cese.to_pickle("data/datiBasePuliti/cese_"+xx+".pickle")
```

```python
def sistema_nome_cognome(str):
    elenco = str.split(sep=" ")
    lung = len(elenco)
    
    diDelList = ["De", "Di", "Del", "Della", "Degli", "La", "Van", "Dal"]
    diDelBool = False
    
    for e in elenco:
        diDelBool = diDelBool | (e in diDelList)
        
    if lung == 2: # Qua normale solo invertire
        return elenco[1] + " " + elenco[0]
    elif diDelBool: # Qua c'√® un Di
        posDi = 0
        for e in elenco:
            if (e not in diDelList):
                posDi = posDi + 1
            else:
                break
        cognome = ""
        nome = ""
        for i in range(posDi, lung):
            cognome = cognome + elenco[i] + " "
        for i in range(0, posDi):
            nome = nome + elenco[i] + " " # attenzione che ci sara' uno spazio di troppo
        risultato = cognome + nome
        return risultato[:-1]
    else: # non solo due parole ma senza Di
        cognome = elenco[-1]
        nome = ""
        for i in range(lung-1):
            nome = nome + " " + elenco[i]
        return cognome + nome
```
