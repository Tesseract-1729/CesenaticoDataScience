```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Fives = pd.read_pickle("data/da_usare/Cese_2003-2024_Fives.pickle")
Fives.tail(6)
```

```python
sns.set( style="whitegrid", font_scale=1.5 )

fig, ax = plt.subplots(figsize = (10, 8))

ax.set_facecolor("0")
ax.grid(True, color="#ADD8E6", linewidth=0.5)

# Disegno le linee
x = np.linspace(1, 5, 5)

for i in range( len(Fives) ):
    d = [Fives.at[i, "pt1"], Fives.at[i, "pt2"], Fives.at[i, "pt3"], Fives.at[i, "pt4"], Fives.at[i, "pt5"]]
    plt.plot(x, d, linestyle="-", linewidth=4, color="#ABCDEF", alpha=0.5)

# Titolo assi e titolo generale
ax.set_title("5 XXXXXXXXXXXXXXXXXX", pad=20, fontsize=30, fontweight="bold")
ax.set_xlabel("Classe", fontsize=25, fontweight="normal", labelpad=15)
ax.set_ylabel("Punteggio totale", fontsize=25, fontweight="normal", labelpad=15);
```

```python
sns.set( style="whitegrid", font_scale=1.5 )

fig, ax = plt.subplots(figsize = (10, 8))

ax.set_facecolor("0")
ax.grid(True, color="#ADD8E6", linewidth=0.5)

# Disegno le linee
x = np.linspace(1, 5, 5)

for i in range( len(Fives) ):
    d = [300 - Fives.at[i, "pos1"], 300 - Fives.at[i, "pos2"],
         300 - Fives.at[i, "pos3"], 300 - Fives.at[i, "pos4"],
         300 - Fives.at[i, "pos5"]]
    plt.plot(x, d, linestyle="-", linewidth=2, color="#1E90FF", alpha=0.3)

# Disegno le linee delle medaglie
plt.hlines(275, 0.9, 5.1, linestyle="-", linewidth=4, color="Gold", label="Oro")
plt.hlines(225, 0.9, 5.1, linestyle="-", linewidth=4, color="Silver", label="Argento")
plt.hlines(150, 0.9, 5.1, linestyle="-", linewidth=4, color="#CD7F32", label="Bronzo")

# Legenda
ax.legend(frameon=True, facecolor="#ADD8E6", framealpha=0.6)

# Tacche e numerini sugli assi
def format_func(value, tick_number):
    n = value
    if n == 300:
        return "1"
    elif n == 0:
        return "300"
    elif n == 275:
        return "25"
    elif n == 225:
        return "75"
    elif n == 150:
        return "150"
def format(value, tick_number):
    n = value
    if n == 300:
        return "1"
    else:
        return str( int(300 - n) )

#ax.set_yticklabels([])
ax.yaxis.set_major_locator(plt.MultipleLocator(50))
ax.yaxis.set_minor_locator(plt.MultipleLocator(25))
ax.yaxis.set_major_formatter(plt.FuncFormatter(format))
ax.yaxis.set_minor_formatter(plt.FuncFormatter(format_func))

# Titolo assi e titolo generale
ax.set_title("Carriere dei concorrenti passati 5 volte", pad=20, fontsize=30, fontweight="bold")
ax.set_xlabel("Classe", fontsize=25, fontweight="normal", labelpad=15)
ax.set_ylabel("Posizione in classifica", fontsize=25, fontweight="normal", labelpad=15)

# Testo a margine
ax.text(0.92, 0.36, "Distanza\ninterquartile\n\nClasse 1:\n123\n\nClasse 5:\n37",
        transform=fig.transFigure, ha="left")
ax.text(0.7, 0.15, "< 1.5%\ndei concorrenti\npassa 5 volte",
        color="White", fontsize=15,
        transform=fig.transFigure, ha="left")

# Salvataggio figura
plt.savefig("image/CarriereConcorrentiPassati5Volte.png", bbox_inches="tight", pad_inches=0.3, transparent=False)
plt.savefig("image/CarriereConcorrentiPassati5Volte.svg", bbox_inches="tight", pad_inches=0.3, transparent=False)
```
