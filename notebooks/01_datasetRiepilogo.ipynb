```python
import numpy as np
import pandas as pd
```

```python
Cese = pd.read_pickle("data/datiBasePuliti/validi/Cese_Corretto_2003-2024.pickle")
Cese
```

```python
df_vuoto = pd.DataFrame(columns=["anno", "num", "pt_bronzo", "pt_argento", "pt_oro", "num_bronzo", "num_argento", "num_oro", "num_menzioni"])
df_vuoto
```

```python
df = df_vuoto
```

```python
df["num"] = Cese.groupby("anno").size()
df["anno"] = Cese.groupby("anno").size().index
#df = df.reset_index(drop = True)
```

```python
df
```

```python
df["num_bronzo"] =  Cese[Cese["medaglia"]=="BRONZO"].groupby("anno").size()
df["num_argento"] = Cese[Cese["medaglia"]=="ARGENTO"].groupby("anno").size()
df["num_oro"] = Cese[Cese["medaglia"]=="ORO"].groupby("anno").size()
df
```

```python
df["num_menzioni"] = Cese[Cese["medaglia"]=="MENZIONE"].groupby("anno").size()
```

```python
df
```

```python
Cese[(Cese["anno"]==2010) & (Cese["es6"]==7)]
# Ho controllato che davvero non ci fossero menzioni
```

```python
df.at[2010, "num_menzioni"] = 0
```

```python
df["pt_bronzo"] = Cese[Cese["medaglia"]=="BRONZO"].groupby("anno").min()["tot"]
df
```

```python
df["pt_argento"] = Cese[Cese["medaglia"]=="ARGENTO"].groupby("anno").min()["tot"]
df["pt_oro"] = Cese[Cese["medaglia"]=="ORO"].groupby("anno").min()["tot"]
df
```

```python
df = df.astype(int)
df.dtypes
```

```python
df
```

```python
df = df.reset_index(drop=True)
df
```

```python
df.to_pickle("data/datiBasePuliti/validi/Cese_riepilogo.pickle")
```
