# Rimozione persone di altri anni
- ordinare il dataset
- stilare di nuovo la classifica

```python
import numpy as np
import pandas as pd
```

```python
def pulizia(anno, xx):
    df = pd.read_pickle("data/datiBasePuliti/validi-constranierrori/cese_"+xx+".pickle")
    df = df[ df["anno"] == anno ]
    
    # Ordino i dati per punteggio
    df = df.sort_values("tot", ascending=False)
    # Sistemo l'indice
    df = df.reset_index(drop=True)
    
    # Popolo la colonna della posizione in classifica
    pos = 1
    df.at[0, "posizione"] = pos
    for i in range(1, len(df.index)):
        if df.at[i, "tot"] == df.at[i-1, "tot"]:
            #cese.at[i, "posizione"] = cese.at[i-1, "posizione"]
            df.at[i, "posizione"] = pos
        else:
            df.at[i, "posizione"] = i+1
            pos = i+1

    df = df.astype({"posizione": int})
    
    # Salvo il database su pickle
    df.to_pickle("data/datiBasePuliti/validi/Cese_20"+xx+".pickle")
```

```python
pulizia(2024, "24")
```
