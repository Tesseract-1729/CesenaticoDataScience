1. **La terza è davvero _l'anno più difficile_?**

   No, crescendo è sia più facile partecipare che ottenere buoni piazzamenti.

2. **Come varia la difficoltà della gara nel tempo?**

   Si alternano anni più o meno difficili.  
   Quella del 2024 sembra essere stata una gara particolarmente semplice, oppure erano più preparati i partecipanti.

4. **Esistono strategie per ottenere medaglie?**

   Per l'argento: risolvere 2 esercizi facili e 1 esercizio difficile.

5. **Come si evolve la carriera degli _habitué_?**

   Prevalentemente coronata dalla medaglia d'oro.

6. **Le rivalità tra regioni limitrofe influenzano gli esiti?  
   Quali sono le regioni più forti?**

   Avere vicini forti rende più forti.  
   Regioni migliori: prevalentemente regioni settentrionali, insieme a Lazio e Puglia.
