# Creiamo i grafici delle medaglie in rerlazione ai punteggi

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Cese = pd.read_pickle("data/da_usare/Cese_2003-2024_data.pickle")
Cese.head(3)
```

```python
plt.style.use("dark_background")

fig, ax = plt.subplots( figsize = (5, 5) )
plt.scatter(Cese[Cese["medaglia"]=="ORO"]["pt123"], Cese[Cese["medaglia"]=="ORO"]["pt456"], alpha=0.5,
           s=100, c=Cese[Cese["medaglia"]=="ORO"]["colore"])
ax.set_xlim(-1, 22)
ax.set_ylim(-1, 22)
ax.set_xlabel("Punteggio negli esercizi 1, 2 e 3")
ax.set_ylabel("Punteggio negli esercizi 4, 5 e 6")
ax.set_title("Distribuzione medaglie d'oro")
ax.grid(False);
```

## Ora faccio il grafico specifico per gli argenti

```python
plt.style.use("dark_background")

fig, ax2 = plt.subplots(figsize=(8, 8))

alf=0.15
radius=170

# Inizio il disegno

ax2.scatter(Cese[Cese["medaglia"]=="ARGENTO"]["pt123"], Cese[Cese["medaglia"]=="ARGENTO"]["pt456"], alpha=alf,
           s=radius, c=Cese[Cese["medaglia"]=="ARGENTO"]["colore"])

ax2.set_xlim(-1, 22)
ax2.set_ylim(-1, 22)
ax2.xaxis.set_major_locator(plt.MultipleLocator(7))
ax2.yaxis.set_major_locator(plt.MultipleLocator(7))

ax2.set_xlabel("Punteggio negli esercizi 1, 2 e 3", labelpad=15, fontsize=25)
ax2.set_ylabel("Punteggio negli esercizi 4, 5 e 6", labelpad=15, fontsize=25)
ax2.set_title("Distribuzione medaglie d'argento", pad=20, fontsize=25, fontweight="bold")

ax2.grid(True, color="Grey", linewidth=1)

# Ora mettiamo in evidenza la croce
alpha_croce = 0.4
larg_croce = 3

l_oriz = ax2.hlines(7, 8, 20, linestyle="-", linewidth=larg_croce, color="Yellow", alpha=alpha_croce)
l_oriz.set_capstyle("round")
l_vert = ax2.vlines(14, 2, 11, linestyle="-", linewidth=larg_croce, color="Yellow", alpha=alpha_croce)
l_vert.set_capstyle("round")

ax2.annotate("Strategia d'argento:\n2 esercizi tra i primi\n1 tra gli ultimi", xy=(14, 7), xytext=(1, 16), 
            arrowprops=dict(edgecolor="Yellow", facecolor="0.5", linewidth=1, shrink=0.04),
            fontsize=20);

# Salvataggio figura
plt.savefig("image/DistribuzioneArgenti.png", bbox_inches="tight", pad_inches=0.3, transparent=False)
plt.savefig("image/DistribuzioneArgenti.svg", bbox_inches="tight", pad_inches=0.3, transparent=False)
```
